```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from datetime import datetime, timedelta

import re
import time
import requests
import traceback
import pandas as pd
from bs4 import BeautifulSoup
```

### Implicit Waits(암묵적 대기)

```python
url = 'https://finance.naver.com/news/news_read.naver?article_id=0001657089&office_id=057&mode=LSS2D&type=0&section_id=101&section_id2=258&section_id3=&date=20220418&page=1'
```

```python
s = Service('./chromedriver')
driver = webdriver.Chrome(service=s)

driver.implicitly_wait(time_to_wait=5)
driver.get(url)
search_content = driver.find_element(by=By.XPATH, value='//*[@id="content"]')
article_date = driver.find_element(by=By.XPATH, value='//*[@id="contentarea_left"]/div[2]/div[1]/div[2]/div/span')
print(article_date.text)
print(search_content.text.replace('\n', ''))

driver.close()
```

```python
class NaverNews(object):
    
    def __init__(self):
        self.base_url = 'https://finance.naver.com/news/news_list.naver'
    def __repr__(self):
        return f'<NaverNew> {self.base_url}'
        
    def get_article_link(self, date_str:str):
        """
        date_str : 'YYYYmmdd' format supported
        """
        
        title_list = []
        article_link = []        
        try:
            response = requests.get(f'{self.base_url}?date={date_str}')
            
            soup = BeautifulSoup(response.text, 'lxml')
            max_page = soup.select_one('td.pgRR > a')
            max_page_href = max_page.attrs['href']
            params = {k:v for k,v in re.findall('(\w+)=(\w+)',max_page_href)}
            
            max_page = int(params['page'])
            
            for page in range(1, max_page+1):
                requests.get(f'{self.base_url}?date={date_str}&page={page}')
                soup = BeautifulSoup(response.text, 'lxml')

                for el in soup.select('#contentarea_left > ul.realtimeNewsList .articleSubject > a'):
                    title_list.append(el.attrs['title'])
                    article_link.append('https://finance.naver.com' + el.attrs['href'])
                    
        except Exception as e:
            traceback.print_exc()
            return [], []
        else:
            return title_list, article_link
```

```python

class CheckElapsedTime():
    
    @staticmethod
    def decorator(function):
        
        def wrapper(*args, **kwargs):
            start_time = time.perf_counter()
            ret = function(*args, **kwargs)
            elapsed_time = time.perf_counter() - start_time
            print(f'elapsed time = {elapsed_time:.2f}')
            return ret
        
        return wrapper
        


class ChromWrapper(object):
    
    def __init__(self):
        self.driver = self.get_driver()
        
    def get_driver(self):
        s = Service('./chromedriver')
        options = webdriver.ChromeOptions()
        options.add_argument('headless')
        options.add_argument('disable-gpu')

        options.add_argument('disable-infobars')
        options.add_argument('--disable-extensions')
        
        driver = webdriver.Chrome(service=s, options=options)
        driver.implicitly_wait(time_to_wait=5)
        
        return driver
    
    @CheckElapsedTime.decorator
    def parse_naver_news(self, link: list):
        
        date_list = []
        content_list = []
        
        for url in link:
            self.driver.get(url)
            search_content = self.driver.find_element(by=By.XPATH, value='//*[@id="content"]')
            article_date = self.driver.find_element(by=By.XPATH, value='//*[@id="contentarea_left"]/div[2]/div[1]/div[2]/div/span')
            
            date_list.append(article_date.text)
            content_list.append(search_content.text)
            
        return date_list, content_list
    
    def __del__(self):
        if isinstance(self.driver,webdriver.Chrome):
            self.driver.close()
```

```python
if __name__ == '__main__':
    news = NaverNews()
    chrome = ChromWrapper()
    title, link = news.get_article_link('20220430')
    date_list, content_list = chrome.parse_naver_news(link)
    df = pd.DataFrame({'title':title, 'date':date_list, 'content': content_list})
```
